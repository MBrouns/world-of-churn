```python
import pandas as pd
from woc import data
```

```python
df = pd.read_csv('../data/wowah_data.csv', nrows=100_000)
df[df['char'] == 59425]
```

```python
clean_df = (
    df
    .pipe(data.normalize_colnames)
    .pipe(data.parse_dtypes)
    .pipe(data.remove_below_min_level, min_lvl=20)
)
```

```python
%%timeit
clean_df.groupby('char')['level'].apply(lambda s: s.max())
```
